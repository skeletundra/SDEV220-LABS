```python
def sort01(arr):
    low = 0
    mid = 0
    high = len(arr) - 1

    while mid <= high:
        if arr[mid] == 0:
            arr[low], arr[mid] = arr[mid], arr[low]
            low += 1
            mid += 1
        elif arr[mid] == 1:
            mid += 1
        else:
            arr[mid], arr[high] = arr[high], arr[mid]
            high -= 1
            
arr = [0, 1, 2, 0, 1, 2, 2, 1, 0, 2, 0, 1, 1, 0, 2]
sort01(arr)
print(arr)
```

```python
class Solution:
    def binarysearch(self, arr, k):
        left = 0
        right = len(arr) - 1
        result = -1

        while left <= right:
            mid = left + (right - left) // 2

            if arr[mid] == k:
                result = mid
                right = mid - 1
            elif arr[mid] < k:
                left = mid + 1
            else:
                right = mid - 1
        return result

sol = Solution()

print(sol.binarysearch([1, 2, 3, 4, 5], 4))
print(sol.binarysearch([11, 22, 33, 44, 55], 445))
print(sol.binarysearch([1, 1, 1, 1, 2], 1))
```
