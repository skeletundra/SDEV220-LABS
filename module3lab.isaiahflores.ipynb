```python
#define vehicle superclass
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

#define subclass automobile
class automobile(Vehicle):
    def __init__(self, year, make, model, doors, roof):
        super().__init__('car')
        self.year = year
        self.make = make
        self.model = model
        self.doors = doors
        self.roof = roof

    #information display
    def display_info(self):
        print(f"vehicle type: {self.vehicle_type}")
        print(f"year: {self.year}")
        print(f"make: {self.make}")
        print(f"model: {self.model}")
        print(f"number of doors: {self.doors}")
        print(f"roof type: {self.roof}")

#get user input, create automobile object
def create_automobile():
    print("please enter your vehicle details:")
    year = input("year: ")
    make = input("make: ")
    model = input("model: ")

    #input validation for doors
    while True:
        doors = input("number of doors (2 or 4): ")
        if doors in['2', '4']:
            doors = int(doors)
            break
        else:
            print("please enter 2 or 4.")

    #input validation for roof
    while True:
        roof = input("type of roof (solid roof or sunroof): ").lower()
        if roof in['solid', 'sunroof']:
            break
        else:
            print("please enter 'solid' or 'sunroof'.")

    car = automobile(year, make, model, doors, roof)
    print("\nhere is your car's information:\n")
    car.display_info()

create_automobile()
```
